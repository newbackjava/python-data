```python
def connect():
    CoffeeBean_URL = "https://www.coffeebeankorea.com/store/store.asp"
    wd = webdriver.Chrome()
    return wd
```

```python
def coffee_bean(no):
    wd.get(CoffeeBean_URL)
    shop = ''
    func = 'storePop2("' + str(no) + '")'
    try:
        time.sleep(1)
        wd.execute_script(func)
        time.sleep(1)
        html = wd.page_source
    #     print(html)
        soupCB = BeautifulSoup(html, 'html.parser')
        # #matizCoverLayer0Content > div > div > div.store_txt > h2
        store_name_h2 = soupCB.select("div.store_txt > h2")
        print('1---', store_name_h2)
        store_name = store_name_h2[0].string
        print('2---', store_name)  #매장 이름 출력하기
        store_info = soupCB.select("div.store_txt > table.store_table > tbody > tr > td")
        print('3---', store_info)
        store_address = store_info[2].getText()
        print('4---', store_address)
        store_phone = store_info[3].getText()
        print('5---', store_phone)
        #shop = store_name + ',' + store_address + ',' + store_phone
        shop = []
        shop.append(store_name)
        shop.append(store_address)
        shop.append(store_phone)
        print('크롤링 결과>>>', shop)
    except:
        print('해당번호없음')
    return shop
```

```python
wd = None
wd = connect()
shop = coffee_bean('2') #특정번호가 없음.
print(shop)
```

```python
wd = None
wd = connect()
shop = coffee_bean(1) #특정번호가 없음.
print(shop)
```

```python
result2 = []
wd = None
wd = connect()
for no in range(1, 20):
    print('=========' + str(no) + '===========')
    shop = coffee_bean(no)
    print(shop)
    if shop != '':
        result2.append(shop)
        print('length>> ', len(result2))
```

```python
CB_tbl = pd.DataFrame(result2,
                      columns=('store', 'address','phone'))
CB_tbl.to_csv('../csv-data/CoffeeBean.csv',
                  encoding='cp949', mode='w', index=True)
```
