```python
class Car: #틀
    # 이니셜라이저 함수 --> 객체(부품)생성시 속성의 처음값(초기값)을 넣고 싶을 때 사용
    def __init__(self, color, speed):
        self.color = color
        self.speed = speed
        
    def __str__(self):
        return self.color + str(self.speed)
        
    def run(self):
        print("달리자~~~~~~~~~~~~")
    
```

```python
car1 = Car("빨강", 150)
car1
```

```python
print(car1)
```

```python
car1.run()
```
