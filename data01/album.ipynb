```python
import pandas as pd
```

```python
album = pd.read_csv("album.csv", encoding="utf8")
album
```

```python
 ## brand 항목 추출해서 프린트
album['brand']
```

```python
 ## brand 항목 추출해서 brand 변수에 넣어보세요
brand =  album['brand']
brand
```

```python
## brand 변수 타입 확인
type(brand)
```

```python
 ## brand에서 값만 추출해서 brand_values넣어 프린트
```

```python
brand_values = brand.values
brand_values
```

```python
## brand_values 변수 타입 확인
type(brand_values)
```

```python
## brand_values를 list()로 바꾸어 for문 돌려 추출
brand_values_list = list(brand_values)
brand_values_list
```

```python
for x in brand_values_list:
    print(x)
```
