```python
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# ===== 1. Titanic: 생존 예측 (로컬 주피터 버전) =====

# 1) 데이터 불러오기
titanic = sns.load_dataset('titanic')

# 2) 숫자 피처 + 타깃 선택
feature_cols_titanic = ['pclass', 'age', 'sibsp', 'parch', 'fare']
target_col_titanic = 'survived'

data_titanic = titanic[feature_cols_titanic + [target_col_titanic]].dropna()
X_titanic = data_titanic[feature_cols_titanic]
y_titanic = data_titanic[target_col_titanic]

# 3) train / test 분리
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(
    X_titanic, y_titanic,
    test_size=0.2,
    stratify=y_titanic,
    random_state=42
)

# 4) 모델 학습
model_titanic = LogisticRegression(max_iter=1000)
model_titanic.fit(X_train_t, y_train_t)

# 5) 예측 및 정확도
y_pred_t = model_titanic.predict(X_test_t)
acc_t = accuracy_score(y_test_t, y_pred_t)
print("Titanic 테스트 정확도:", round(acc_t, 3))

# 6) 리스트로 새 승객 예측
sample_titanic = [[3, 22, 1, 0, 7.25]]  # [pclass, age, sibsp, parch, fare]
pred_sample_t = model_titanic.predict(sample_titanic)[0]
print("예시 승객 생존 예측(0=사망, 1=생존):", pred_sample_t)
```
