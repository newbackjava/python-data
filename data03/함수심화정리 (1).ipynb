````python
# 입력값이 여러개를 넣고 싶을 때 함수 정의
# 가변 매개변수
def print_fruits(*names): # 입력한 값들의 목록은 무조건 tuple(읽기 전용)
  print(names) #튜플 : ('사과', '배')
  for x in names:
    print("내가 좋아하는 과일은 ", x)
````

````python
print_fruits('apple')
````

````python
print_fruits('사과', '배')
````

````python
print_fruits('사과', '배', '딸기')
````

````python
# 키역할을 할 수 있는 "키워드 = 값"형태로 넣을 수 있음.
# 키워드 가변 매개변수
def comment_info(**likes): # 딕셔너리로 인식
  print(likes) # {'fruit': '사과', 'drink': '사이다'}
  for key, value in likes.items():
    print(key, ": ", value)
````

````python
comment_info(fruit="사과", drink="사이다")
````

람다(lambda)함수

````python
def call1(a):
  return a**2 # a * a
````

````python
print(call1(10))
````

````python
print((lambda a: a**2)(10))
````

````python
# df의 컬럼에 값들을 여러 복잡한 처리를 해야하는 경우
# 함수를 이용해서 처리를 함.
````

````python
a = [1, 2, 3]
result = map(call1, a) #map(처리할 함수, 반복할리스트)
print(list(result)) #map처리할 결과는 반드시 list()로 바꿔주세요.!!!
````

````python
a2 = [10, 20, 30]
result2 = map(lambda a: a**3, a2)
print(list(result2))
````

````python
import pandas as pd
import numpy as np
````

````python
df1 = pd.DataFrame({
    'name': ['홍길동', '김길동', '홍순신', '정길동', '송사임'],
    'age': [100, 200, 300, 150, 250]
})
df1
````

````python
df1['result'] = ['양수', '양수', '음수', '음수', '양수']
````

````python
df1
````

````python
# 1. age컬럼 추출해서 반복해서 하나씩 꺼내야한다.
#     --> map(함수명, 반복할처리할 컬럼)
#     --> 컬러명.apply(함수명), for
# 2. 조건 처리, 200살 이상이 맞으면 "노령", 아니면 "보통" --> 일반함수, 람다함수
# 3. age_result컬럼을 새로 파생변수로 2번에서 처리한 리스트를 넣을 예정임.
#    --> map()의 결과는 Map객체로 리턴을 해줌.
#    --> list(map())로 바꾸어주자.
````

````python
list(map(lambda x : "처리결과", df1['age']))
````

````python
list(map(lambda x : "고령" if x >=200 else "보통", df1['age']))
````

````python
df1['age_result'] = list(map(lambda x : "고령" if x >=200 else "보통", df1['age']))
df1
````

````python
# 1. name컬럼 추출
# 2. name컬럼값을 하나씩 꺼내어
# 3. 참석자명단에 그 이름이 있으면 "참석함", 없으면 "참석하지 않음."
# ["홍길동", "홍순신", "송사임"]
# 4. name_result변수에 결과를 넣음.
````

````python
a = ["홍길동", "홍순신", "송사임"]
print(a[0] == '홍길동')
````

````python
print(a[0] in ['홍길동', '송사임'])
````

````python
# map(함수명 또는 람다, 반복처리대상)
list(map(lambda x : x + "처리내용", df1['name']))
````

````python
list(map(lambda x : "참석함" if x in ["홍길동", "홍순신", "송사임"] else "참석하지않음", df1['name']))
````

````python
df1['name_result'] = list(map(lambda x : "참석함" if x in ["홍길동", "홍순신", "송사임"] else "참석하지않음", df1['name']))
df1
````
