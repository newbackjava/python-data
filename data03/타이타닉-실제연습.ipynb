```python
# 컬럼의 값을 하나씩 꺼내서 처리할 함수
# 컬럼의 값이 3개인 경우 이 함수는 3번 호출됨.
def pclass_convert(pclass):
    if pclass == 1:
        return "low"
    elif 2 <= pclass <= 3: # 2 <= pclass and pclass <= 3
        return "middle"
    else:
        return "high"
```

```python
# df['pclass'] = df['pclass'].apply(적용할 함수)
df['pclass'] = df['pclass'].apply(pclass_convert) #apply의 list(Series)
df
```

```python
df['pclass'].value_counts() #값별로 카운트를 세줌.
```

```python
def ages(age):
    if age < 10:
        return "유아"
    elif 10 <= age < 20:
        return "10대"
    elif 20 <= age < 30:
        return "20대"
    elif 30 <= age < 40:
        return "30대"
    elif 40 <= age < 50:
        return "40대" 
    elif 50 <= age < 60:
        return "50대" 
    else:
        return "60대이상"
```

```python
df['age'] = df['age'].apply(ages)
df['age']
```

```python
df['age'].value_counts()
```

```python
result_count = df['age'].value_counts()
type(result_count)
```

```python
def name_gender(name) :
    if 'Mr' in name or 'Dr' in name:
        return "남자"
    elif 'Miss' in name:
        return "여자"
    else:
        return "기타"
```
