```python
# 행추출 2개를 많이 씀.
# df.loc() --> 현재 인덱스로 잡혀있는 컬럼의 값으로 행 추출
# df.iloc() --> 현재 인덱스로 잡혀있는 컬럼의 인덱스값으로 행 추출
```

```python
df
```

```python
# 데이터프레임을 만들 때 인덱스를 지정안하면 정수자동증가기능을 이용해서 
# 값이 정수값으로 이루어진 정수 인덱스가 생성됨.
# 값인덱스 [0,1,2,3,4,5] 생성 --> loc[]
# 위치인덱스 [0,1,2,3,4,5] 생성 --> iloc[]

# 데이터프레임을 만들 때 인덱스를 지정하면 내가 지정한 값으로 인덱스가 생성됨.
# apple, ice가 인덱스를 생성한 경우 문자열로 인덱스가 생성됨.
# 이때의 경우에는 숫자 인덱스를 쓸 수 있음.
# 값인덱스 ["apple", "ice"] 생성 --> loc[]
# 위치인덱스 [0,1] 생성 --> iloc[]
```
